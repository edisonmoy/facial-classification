### Facial Detection

Using OpenCV, this basic model can detect faces from uploaded images.

```python
import cv2 as cv
from matplotlib import pyplot as plt

#Image path
path = 'friends.jpg'

#Import image
img = cv.imread(path)

plt.imshow(img)
plt.title('Original Photo')
plt.show()
```

```python
#Convert to grayscale

gs_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(gs_img)
plt.title('Grayscale Photo')
plt.show()
```

```python
#Import and configure classifier cascade

face_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')

detected_faces = face_cascade.detectMultiScale(gs_img)
```

```python
#Draw rectangles around detected faces

for (column, row, width, height) in detected_faces:
    cv.rectangle(
        img,
        (column, row),
        (column + width, row + height),
        (255, 0, 0),
        2
    )
    
plt.imshow(img)
plt.title('Faces detected ')
plt.show()
```
